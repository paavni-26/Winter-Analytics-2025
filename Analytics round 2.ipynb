```python
import pandas as pd
import matplotlib.pyplot as plt


df = pd.read_csv('china data.csv') 
df.columns = [col[1] if isinstance(col, tuple) else col for col in df.columns]

# Extracting 2005 and 2015 data
data_2005 = df.filter(like="2005 [YR2005]").rename(lambda x: x.replace("2005 [YR2005]", "").strip(), axis=1)
data_2015 = df.filter(like="2015 [YR2015]").rename(lambda x: x.replace("2015 [YR2015]", "").strip(), axis=1)

# Ensuring numeric values
data_2005 = data_2005.apply(pd.to_numeric, errors="coerce")
data_2015 = data_2015.apply(pd.to_numeric, errors="coerce")

# Compute change (2015 - 2005)
change = (data_2015.iloc[0] - data_2005.iloc[0]).sort_values(ascending=False)

# Select top 10 positive impact indicators
top_positive_indicators = change.head(10)

#top 10 indicators
plt.figure(figsize=(12, 6))
top_positive_indicators.plot(kind="barh", color="skyblue")

plt.xlabel("Change from 2005 to 2015")
plt.ylabel("Indicators")
plt.title("Top 10 Indicators Positively Impacting China's Development (2005-2015)")
plt.gca().invert_yaxis()

plt.show()
```

What is Correlation?
+1 → Strong positive correlation (both increase together).
-1 → Strong negative correlation (one increases, the other decreases).

```python
df = net_zero_analysis
def plot_progress(series_code, title, ylabel):
    """Plot progress of a specific indicator."""
    data = df[df['Series Code'] == series_code].iloc[:, 4:]
    
    if data.empty:
        print(f"Warning: {series_code} not found in dataset.")
        return
    
    data = data.squeeze()  
    data.index = [int(year.split(' ')[0]) for year in data.index]
    
    plt.figure(figsize=(8, 5))
    plt.plot(data.index, data.values, marker='o', label=title)
    plt.title(title)
    plt.xlabel("Year")
    plt.ylabel(ylabel)
    plt.grid()
    plt.legend()
    plt.show()
indicators = {
    "EG.ELC.RNWX.ZS": "Electricity production from renewable sources, excluding hydroelectric (% of total)",
    "EG.ELC.COAL.ZS": "Electricity production from coal sources (% of total)",
    "EG.USE.ELEC.KH.PC": "Electric power consumption (kWh per capita)",
    "EN.ATM.PM25.MC.M3": "PM2.5 air pollution, mean annual exposure (micrograms per cubic meter)"

}

for code, title in indicators.items():
    if code in df["Series Code"].values:  
        plot_progress(code, title, title)
    else:
        print(f"Warning: {code} not found in dataset.")
```

Key Point to Track: India aims to meet 50% of total energy requirements through renewable sources by 2030. The trajectory should continue rising at a faster rate to meet this goal.  

ii)Electricity production from coal sources (% of total)
The second graph shows fluctuations, but overall, coal-based electricity remains a dominant source.
Key Point to Track: India aims to reduce carbon intensity by 45% from 2005 levels. A significant decline in coal dependency is needed.

iii)Electric power consumption (kWh per capita):
This graph shows a rising trend in power consumption per capita.
Key Point to Track: While energy consumption is growing, the share from renewables must grow proportionally to align with Panchamrit commitments.

iv)There is a sharp peak around 2010-2011, followed by a steep decline in 2012.
Long-Term Trend: Despite fluctuations, is there an overall decline in PM2.5 levels?

```python
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.arima.model import ARIMA
df = pd.read_csv('Book1.csv')
 

melted_df = df.melt(id_vars=[df.columns[0]], var_name="Year", value_name="Value")
melted_df.columns = ["Indicator", "Year", "Value"]


melted_df["Year"] = melted_df["Year"].str.extract("(\\d+)").astype(int)  

indicator_to_analyze = melted_df["Indicator"].iloc[0]
ts_data = melted_df[melted_df["Indicator"] == indicator_to_analyze].set_index("Year")["Value"] 

# Fit ARIMA model
model = ARIMA(ts_data, order=(2,0,0))  # Adjust parameters as needed
model_fit = model.fit()

# Forecast future values
future_years = list(range(ts_data.index[-1] + 1, 2071))
forecast = model_fit.forecast(steps=len(future_years))

# Plotting original data and forecast
plt.figure(figsize=(10, 5))
plt.plot(ts_data.index, ts_data, label="Actual Data", marker="o")
plt.plot(future_years, forecast, label="Forecast", linestyle="dashed", marker="o")
plt.xlabel("Year")
plt.ylabel("Value")
plt.title(f"Forecast for {indicator_to_analyze}")
plt.legend()
plt.grid()
plt.show()   
```

The blue points represent actual data for the share of renewable energy in final energy consumption.
The orange points indicate the forecasted trend for the future.

Key Observations:
The share of renewables declined until around 2015, reaching its lowest point.
After 2020, the forecast shows a steady increase in renewable energy consumption, peaking around 2040 at nearly 42%.
However, a fluctuating trend is observed beyond 2040, with periodic declines and recoveries, stabilizing around 38% by 2070.
Renewable energy consumption is projected to rise steadily until 2040 but fluctuates afterward, stabilizing at around 38% by 2070.

Implications:
Renewable energy is expected to grow but with instability, indicating policy or economic fluctuations affecting adoption.
Sustained investments, policy consistency, and grid modernization are needed to ensure a smooth and consistent renewable energy transition.
The overall increase suggests a positive shift toward sustainable energy but not at an aggressive pace needed for net-zero goals.
